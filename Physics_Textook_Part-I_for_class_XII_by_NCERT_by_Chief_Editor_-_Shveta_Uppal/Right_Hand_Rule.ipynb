# Right Hand Thumb Rule

```python
# Right Hand Thumb Rule
# Reference: University of Redlands:Computational Tutorials
%matplotlib inline
from pylab import *
# Set limits and number of points in grid
# value of xmax and ymax can be changed to change the direction of magnetic field lines and can have any value except zero
xmax = 4
xmin = -xmax
NX = 10
ymax = 4
ymin = -ymax
NY = 10
# Make grid and calculate vector components
x = linspace(xmin, xmax, NX)
y = linspace(ymin, ymax, NY)
X, Y = meshgrid(x, y)
S2 = X**2 + Y**2 # This is the radius squared
Bx , By= -Y/S2 , +X/S2
if(xmax>0 or ymax>0):
    plot(0,0,'ro')                   # Red Dot to represent direction of current is out of the plane
    QP = streamplot(X,Y,Bx,By,density=1.6)
else:
	plot(0,0,'go')                   # Green Dot to represent direction of current is into the plane
	QP = streamplot(-X,-Y,Bx,By,density=1.6)
# Set the left, right, bottom, top limits of axes
title('Magnetic Field for current')
xlabel('x (cm)'), ylabel('y (cm)')
show()
```
